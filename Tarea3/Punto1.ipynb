```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def System(r,t):
    x,y = r
    dxdt = 2*x - y
    dydt = x + 2*y
    return np.array([dxdt,dydt])

t = np.linspace(0,3,1000)
r0 = np.array([0.1,0.1])
sol = odeint(System, r0, t)
plt.plot(t,sol[:,0])
```
